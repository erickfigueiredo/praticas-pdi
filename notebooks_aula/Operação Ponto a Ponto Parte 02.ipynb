## **2.1 Função `exibe_mosaico`**

- Para mais detalhes, consulte o Notebook do capítulo "Entropia, Ruídos e Métricas"

```python
def exibe_mosaico(imagens, legendas, colunas=4, arquivo=None):
  # Largura e altura básicos, de referência, das imagens
  l = 8
  a = 6

  n_imagens = len(imagens)
  if len(imagens) < colunas:
    colunas = len(imagens)
  linhas = n_imagens // colunas + int(n_imagens % colunas > 0)

  fig = plt.figure(figsize=(l*colunas,a*linhas))
  ax = []

  i = j = 0
  for pos, img in enumerate(imagens):
    i = pos // colunas
    j = pos % colunas
    ax.append(fig.add_subplot(linhas, colunas, pos+1))
    ax[-1].axis('off')
    if img.mode == 'L' or img.mode == '1':
      img = img.convert('RGB')
    arr = np.asarray(img)
    ax[-1].imshow(arr)
    ax[-1].set_title(legendas[pos])

  fig.tight_layout()

  if arquivo is not None:
    plt.savefig(arquivo, dpi=300)
    
  plt.show()
```

## **2.2 Função `gera_imagem`**

- Para mais detalhes, consulte o Notebook do capítulo "Entropia, Ruídos e Métricas"

```python
def gera_imagem(arr_orig):
  arr = arr_orig.copy()
  arr[arr > 255] = 255
  arr[arr < 0] = 0
  arr = np.round(arr).astype(np.uint8)
  return Image.fromarray(arr)
```

## **3.4 Conversão de espaço de cores**

```python
def rgb2hsl(arr):
  hsl = np.zeros_like(arr).astype(float)

  arrNorm = arr/255.0

  Cmax = np.max(arrNorm, axis=2)
  iCmax = np.argmax(arrNorm, axis=2)
  
  Cmin = np.min(arrNorm, axis=2)

  delta = Cmax - Cmin

  L = (Cmax + Cmin) / 2

  H = np.empty(hsl[:,:,0].shape)

  H[delta == 0] = 0

  calculo = H.copy()
  calculo[delta != 0] = 60 * (((arrNorm[:,:,1][delta != 0] - arrNorm[:,:,2][delta != 0]) / delta[delta != 0]) % 6)
  H[iCmax == 0] = calculo[iCmax == 0]

  calculo = H.copy()
  calculo[delta != 0] = 60 * (((arrNorm[:,:,2][delta != 0] - arrNorm[:,:,0][delta != 0]) / delta[delta != 0]) + 2)
  H[iCmax == 1] = calculo[iCmax == 1]

  calculo = H.copy()
  calculo[delta != 0] = 60 * (((arrNorm[:,:,0][delta != 0] - arrNorm[:,:,1][delta != 0]) / delta[delta != 0]) + 4)
  H[iCmax == 2] = calculo[iCmax == 2]
  
  hsl[:,:,0] = H.copy()

  S = np.zeros(hsl[:,:,1].shape)
  S[delta != 0] = delta[delta != 0] / (1 - np.abs(2 * L[delta != 0] - 1))
  hsl[:,:,1] = S.copy()

  hsl[:,:,2] = L.copy()

  return hsl
```

```python
def hsl2rgb(arr):
  rgb = np.zeros_like(arr).astype(float)
  zerada = np.zeros(arr[:,:,0].shape)

  C = (1 - np.abs(2 * arr[:,:,2] - 1)) * arr[:,:,1]
  X = C * (1 - np.abs(arr[:,:,0]/60 % 2 - 1))
  m = arr[:,:,2] - C / 2
  m = np.dstack((m,m,m))
  
  h = arr[:,:,0]
  
  cond = (h >= 0) & (h < 60)
  rgb[cond] = np.dstack((C[cond], X[cond], zerada[cond]))

  cond = (h >= 60) & (h < 120)
  rgb[cond] = np.dstack((X[cond], C[cond], zerada[cond]))

  cond = (h >= 120) & (h < 180)
  rgb[cond] = np.dstack((zerada[cond], C[cond], X[cond]))

  cond = (h >= 180) & (h < 240)
  rgb[cond] = np.dstack((zerada[cond], X[cond], C[cond]))

  cond = (h >= 240) & (h < 300)
  rgb[cond] = np.dstack((X[cond], zerada[cond], C[cond]))

  cond = (h >= 300) & (h <= 360)
  rgb[cond] = np.dstack((C[cond], zerada[cond], X[cond]))

  rgb = np.round((rgb + m) * 255)

  return rgb.astype(int)
```

```python
img = Image.open(path + 'ipe_amarelo.jpg')
arr = np.asarray(img).astype(float)
arrHSL = rgb2hsl(arr)
arrRGB = hsl2rgb(arrHSL)

rmse = np.sqrt(np.sum((arr - arrRGB)**2))
print('RMSE = %.2f' % rmse)

fator = 0.5
arrHSLh = arrHSL.copy()
arrHSLh[:,:,0] *= fator
arrRGBh = hsl2rgb(arrHSLh)

arrHSLs = arrHSL.copy()
arrHSLs[:,:,1] *= fator
arrRGBs = hsl2rgb(arrHSLs)

arrHSLl = arrHSL.copy()
arrHSLl[:,:,2] *= fator
arrRGBl = hsl2rgb(arrHSLl)

imagens = [img, gera_imagem(arrRGB), gera_imagem(arrRGBh), gera_imagem(arrRGBs), gera_imagem(arrRGBl)]
legendas = ['Original', 'Reconstruída', 'Alterada no canal H (matiz)', 'Alterada no canal S (saturação)', 'Alterada no canal L (luminosidade)']
exibe_mosaico(imagens, legendas)
```

```python
def rgb2yiq(arr):
  T = [
      [0.299, 0.587, 0.114],
      [0.596, -0.275, -0.321],
      [0.212, -0.523, 0.311]
  ]
  T = np.array(T).T
  return np.dot(arr, T)

def yiq2rgb(arr):
  T = [
      [0.299, 0.587, 0.114],
      [0.596, -0.275, -0.321],
      [0.212, -0.523, 0.311]
  ]
  T = np.linalg.inv(np.array(T).T)
  return np.dot(arr, T)
```
